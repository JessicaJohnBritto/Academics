```python
# x1 = x +0.01
# X = [0.09, 0.08]
def CAc(A):
    x, y = [], []
    xn = 0.9
    xnn = 0.91
    for n in range(100):
        xn1=A*xn*(1-xn)
        xn=xn1
        xn2 = A*xnn*(1-xnn)
        xnn = xn2
        x.append(abs(xn-xnn)/0.01)
        y.append(n)
    plt.plot(y,x)
    plt.xscale('log')
#     plt.yscale('log')

CAc(2.9)
CAc(3.0)
CAc(3.1)
```

# Root finding 1

```python
## Find number of real roots and roots itself
def f(x):
    return ((2*x**3) -5.0)
```

```python
X=np.arange(-5,5,0.01)
Y=f(X)
```

```python
plt.plot(X, Y)
```

```python
e = 0.001
def k(a,b):
    t=(a+b)/2
    if ((f(a)==0) or f(b)==0):
        if f(a)==0:
            return a
        elif f(b)==0:
            return b
    elif (f(a)*f(b))>0:
        return 0
    else:
        if abs(f(t))>e:
            if f(t)==0:
                return t
            elif not(f(t)==0):
                if f(a)*f(t)<0:
                    b=t
                    return k(a,b)
                elif f(b)*f(t)<0:
                    a=t
                    return k(a,b)
        elif abs(f(t))<e:
            return t

g=k(-1,3)
if g==0:
    print("Choose inputs such that a<b and the product of their outputs is less than zero\n")
else:
    print(g)
print(f(g))
```

# Root finding 2

```python
def g(x):
    return ((x**3)-(0.165*(x**2))+(3.993*(10**(-4))))
```

```python
def h(x):
    return ((3*(x**2))-(2*0.165*(x**1)))
```

```python
e=0.0001
x=0.05
i=0
while abs(g(x))>e:
    x=(x)-((g(x))/(h(x)))
print(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

# ODE-1

```python
y=4
h=0.05
N=60

def f(x,y):
    return (((2*y)-(18*x))/(1+x))

def euler(x, y, h, N):
    e=np.linspace(0,3, num=N)
    X, Y = [], []
    for i in e:
        y_new = y + (h*(f(i,y)))
        y = y_new
        X.append(i)
        Y.append(y)
    return plt.plot(X,Y, color="red")

def analytic_function(x):
    return ((-5*(x**2))+(8*x)+4)

def analytical(N):
    e=np.linspace(0,3, num=N)
    X, Y = [], []
    for i in e:
        X.append(i)
        Y.append(analytic_function(i))
    return plt.plot(X,Y, color="blue")

# Rung-Kutta Method 2nd order
def rk2(x, y, h, N):
    e=np.linspace(0,3, num=N)
    X, Y = [], []
    for i in e:
        k1 = h*f(i,y)
        k2 = h*f(i+h/2, y+k1/2)
        y_new = y + k2
        y = y_new 
        X.append(i)
        Y.append(y)
    return plt.plot(X,Y, color="green")
```

```python
analytical(60)
euler(0, 4, 0.05, 60)
rk2(0,4,0.05,60)
```

```python
def error(N):
    e=np.linspace(0,3, num=N)
    X, Y =[],[]
    y=4
    for i in e:
        y_n = y + (h*(f(i,y)))
        y = y_n
        y_a = analytic_function(i)
        error = (abs(y_a-y_n))
        X.append(i)
        Y.append(error)
    return plt.plot(X,Y, color="purple")
error(60)
```

# Coupled ODE - 3

```python
def gg(x,p):
    m=1
    return (p/m)
def hh(x,p):
    k=1
    return (-k*x)
time = np.arange(1,6, 0.01)
def C_ODE(X, t):
    p=X[1]
    x=X[0]
    h=0.05
    X, Y, L = [], [], []
    for i in time:
        p_n = p + t*(gg(x,p))
        x_n = x + t*(hh(x,p))
        p = p_n
        x = x_n
        X.append(x)
        Y.append(p)
    return plt.plot(X,Y, color="orange")
X = [1.,0.1]
C_ODE(X, 0.01)
```

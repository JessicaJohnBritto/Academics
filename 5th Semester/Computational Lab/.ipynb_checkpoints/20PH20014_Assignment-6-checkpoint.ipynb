```python
import numpy as np
import matplotlib.pyplot as plt
```

# ODE-1

```python
y=4
h=0.05
N=60

def f(x,y):
    return (((2*y)-(18*x))/(1+x))

def euler(x, y, h, N):
    e=np.linspace(0,3, num=N)
    X, Y = [], []
    for i in e:
        y_new = y + (h*(f(i,y)))
        y = y_new
        X.append(i)
        Y.append(y)
    return plt.plot(X,Y, color="red", label = "Euler")

def analytic_function(x):
    return ((-5*(x**2))+(8*x)+4)

def analytical(x,h,N):
    e=np.linspace(x, x+h*N, N+1)
    X, Y = [], []
    for i in e:
        X.append(i)
        Y.append(analytic_function(i))
    plt.plot(X,Y, color="blue", label = "Analytical")
    return(Y)

# Rung-Kutta Method 2nd order
def rk2(x, y, h, N):
    e=np.linspace(0,3, num=N)
    X = []
    Y = []
    for i in e:
        Y.append(y)
        k1 = h*f(i,y)
        k2 = h*f(i+ h*0.5, y+ k1*0.5)
        y_new = y + k2
        y = y_new 
        X.append(analytic_function(i))
    plt.plot(e,Y, color="green", label = "RK-2")
    return Y, X, e
```

```python
analytical(0, 0.05, 60)
plt.legend(loc='best')
euler(0, 4, 0.05, 60)
plt.legend(loc='best')
Y, X, e=rk2(0,4,0.05,60)
plt.legend(loc='best')
```

```python
# Runge-Kutta Error
def error1(Y, X, e):
    E = []
    for i in range(len(X)):
        E.append(abs(X[i]-Y[i]))
    return plt.plot(e, E, color = "blue", label = "Runge-Kutta")

error1(Y, X, e)
plt.legend(loc='best')
# Euler
def error(N):
    h=0.05
    e=np.linspace(0,3, num=N)
    X, Y =[],[]
    y=4
    for i in e:
        y_n = y + (h*(f(i,y)))
        y = y_n
        y_a = analytic_function(i)
        error = (abs(y_a-y_n))
        X.append(i)
        Y.append(error)
    return plt.plot(X,Y, color="purple", label = "Euler's")
error(60)
plt.legend(loc='best')
```

# Coupled ODE - 1

```python
# Part - 1

def gg(x,p):
    m=1
    return (p/m)
def hh(x,p):
    k=1
    return (-k*x)
def C_ODE(X, t, hh, gg):
    p=X[1]
    x=X[0]
    time = np.arange(0,20*(np.pi), t)
    XX, Y, L = [], [], []
    for i in time:
        p_n = p + t*(hh(x,p))
        x_n = x + t*(gg(x,p))
        p = p_n
        x = x_n
        XX.append(x)
        Y.append(p)
    plt.plot(XX,Y, label="p vs x")
    plt.legend(loc='best')
    return XX,Y, time
X = [1.,0.1]
```

```python
XX1, Y1, t1 = C_ODE(X, 2*(np.pi)/10, hh, gg)
plt.figure()
plt.plot(t1, XX1, color = "orange", label = "x vs time")
plt.legend(loc='best')
plt.plot(t1, Y1, color = "yellow", label = "p vs time")
plt.legend(loc='best')
```

```python
XX2, Y2, t2 = C_ODE(X, 2*(np.pi)/100, hh, gg)
plt.figure()
plt.plot(t2, XX2, color = "orange", label = "x vs time")
plt.legend(loc='best')
plt.plot(t2, Y2, color = "yellow", label = "p vs time")
plt.legend(loc='best')
```

```python
# Analytical Values of x and p from equation - x^2 + p^2 = c
# Parameterising circle

def analytical_xp(x,p, t):
    c = (p**2) + (x**2)
    return (c*np.cos(t), c*np.sin(t))

def C_ODE2(X, t, analytical_xp):
    p=X[1]
    x=X[0]
    time = np.arange(0,20*(np.pi), t)
    XX, Y, L = [], [], []
    for i in time:
        l = analytical_xp(x, p, i)
        XX.append(l[0])
        Y.append(l[1])
    plt.plot(XX,Y, label = "p vs x")
    plt.legend(loc='best')
    return XX,Y, time
```

```python
X = [1.,0.1]
XX, Y, t = C_ODE2(X, 2*(np.pi)/10, analytical_xp)
plt.plot(XX, Y)
plt.figure()
plt.plot(t, XX, color = "orange", label = "x vs time")
plt.legend(loc='best')
plt.plot(t, Y, color = "yellow", label = "p vs time")
plt.legend(loc='best')
```

```python
XX, Y, t = C_ODE2(X, 2*(np.pi)/100, analytical_xp)
plt.plot(XX, Y)
plt.figure()
plt.plot(t, XX, color = "orange", label = "x vs time")
plt.legend(loc='best')
plt.plot(t, Y, color = "yellow", label = "p vs time")
plt.legend(loc='best')
```

```python
# Part - b

# Plotting V vs x
def pot_x(g1, g2):
    X, Y =[], []
    xx = np.arange(g1, g2, 0.01)
    for i in xx:
        V = ((i**2) -1)**2
        Y.append(V)
    return plt.plot(xx, Y)

pot_x(-2,2)
```

```python
# dx/dt = p/m, dp/dt = - 4x(x^2 - 1)

def pp(x,p):
    return -4*x*((x**2)-1)

def gg(x,p):
    m = 1
    return p

def C3(X, t, pp, gg):
    p=X[1]
    x=X[0]
    time = np.arange(0,5, t)
    XX, Y = [], []
    for i in time:
        p_n = p + t*(pp(x,p))
        x_n = x + t*(gg(x,p))
        p = p_n
        x = x_n
        XX.append(x)
        Y.append(p)
#         plt.plot(XX,Y)
    return XX,Y, time

X = [1.,0.1]
XX1, Y1, t1 = C3(X, 0.0005, pp, gg)
plt.plot(XX1, Y1)
plt.figure()
plt.plot(t1, XX1, color = "orange")
plt.plot(t1, Y1, color = "yellow")
```

# Coupled ODE - 2

```python
def F(x,y, M, z):
    return np.array([-x -y*z, -y -x*z])

def RK2(F, x, z0, y0, h, N, M):
    e=np.linspace(x[0],x[1], num=N)
    e = np.array(e)
    X = np.zeros(N)
    Y = np.zeros(N)
    Z = np.zeros(N)
    Y[0] = y0
    Z[0] = z0
    X[0] = e[0]
    for i in range(len(e)-1):
        k1 = h*F(X[i],Y[i], M, Z[i])
        k2 = h*F(X[i]+ h*0.5, Y[i]+ k1[0]*0.5, M,Z[i] + k1[1]*0.5)
        Y[i+1] = Y[i] + k2[0]
        Z[i+1] = Z[i] + k2[1]
        X[i+1] = e[i+1]
    plt.plot(e,Y, color="green", label = "Y vs X")
    plt.legend(loc='best')
    plt.plot(e,Z, color="red", label = "Z vs X")
    plt.legend(loc='best')
    return Y, Z, e
x = [0.0, 0.5]
RK2(F, x, 1, 0, 0.05, 10, 2)
```

# Coupled ODE - 4

```python
def F(E, B, v):
    s = -1.758*(10**11)
    vx = (s)*(E[0] + (v[1]*B[2]) - (v[2]*B[1]))
    vy = (s)*(E[1] - (v[0]*B[2]) + (v[2]*B[0]))
    vz = (s)*(E[2] + (v[0]*B[1]) - (v[1]*B[0]))
    return np.array([vx, vy, vz])

def LRK2(F, E, B, v, N, M, h):
    e=np.linspace(0,10**(-6), num=N)
    e = np.array(e)
    W = np.ndarray(shape=(M,N))
    for k in range(M):
        W[k][0] = v[k]
    for i in range(len(e)-1):
        k1 = h*F(E,B, W[:, i:i+1])
        k2 = h*F(E, B, W[:, i:i+1] + k1*0.5)
        W[:, i+1:i+2] = k2 + W[:, i:i+1] 
#     V = np.zeros(M)
#     for l in range(M):
    Vx = np.reshape(W[0:1, :], (e.shape))
    Vy = np.reshape(W[1:2, :], (e.shape))
    Vz = np.reshape(W[2:3, :], (e.shape))
    plt.plot(e,Vx, color="green")
    plt.plot(e,Vy, color="red")
    plt.plot(e,Vz, color="yellow")
    return Vx, Vy, Vz, e

# Part i when B is 10^-4 T along z-axis

v = np.array([1.0, 1.0, 1.0])
E = np.array([0.0, 0.0, 0.0])
B = np.array([.0, .0, 10**(-4)])
Vx, Vy, Vz, e = LRK2(F, E, B, v, 1000, 3, 10**(-9))
```

```python
def xyz(Vx, Vy, Vz, e):
    x = np.zeros(len(e))
    y = np.zeros(len(e))
    z = np.zeros(len(e))
    x[0] = Vx[0]*e[0]
    y[0] = Vy[0]*e[0]
    z[0] = Vz[0]*e[0]
    for i in range(len(e)-1):
        x[i+1] = x[i] + Vx[i+1]*(e[i+1] - e[i])
        y[i+1] = y[i] + Vy[i+1]*(e[i+1] - e[i])
        z[i+1] = z[i] + Vz[i+1]*(e[i+1] - e[i])
    return x, y, z

x, y, z = xyz(Vx, Vy, Vz, e)

plt.plot(y, x, label="x vs y", color = "green")
plt.legend(loc='best')
plt.plot(z, y, label="y vs z", color = "blue")
plt.legend(loc='best')
plt.plot(x, z, label="z vs x", color = "red")
plt.legend(loc='best')
```

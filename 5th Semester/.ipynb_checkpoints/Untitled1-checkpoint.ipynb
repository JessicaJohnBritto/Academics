```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Euler method
def Euler(x,y,f,h,n):
    x_list = []
    y_list = []
    for i in range(n):
        y = y + h*f(x,y)
        x_list.append(x)
        y_list.append(y)
        x = x+h
    return x_list,y_list

def rk2(x,y,f,h,n):
    x_list = []
    y_list = []
    for i in range(n):
        k1 = h*f(x,y)
        k2 = h*f(x+h/2, y+k1*h/2)
        y = y + k2
        x_list.append(x)
        y_list.append(y)
        x = x+h
    return x_list,y_list

def plot_func(x_list,y_list):
    plt.plot(x_list,y_list)

def f(x,y):
    return (2*y-18*x)/(1+x)
```

```python
x_or = np.arange(0,3,3/60)
y_or = -5*x_or*x_or + 8*x_or + 4
```

```python
x_eul, y_eul = Euler(0,4,f,0.05,60)
x_rk, y_rk = rk2(0,4,f,0.05,60)
```

```python
plot_func(x_or,y_or)   #blue
plot_func(x_eul,y_eul) #orange
plot_func(x_rk,y_rk)   #green
```

```python
#Absolute error
y_err = abs(y_eul - y_or)
plot_func(x_or, y_err)
#y_err = abs(y_rk - y_or)
plot_func(x_rk, y_err)
```

```python
def Euler_gen(X,F,t,n):
    LIST = []
    for i in range(len(X)):
        LIST.append([])        
    for k in range(n):
        for i in range(len(X)):
            X[i] = X[i] + F[i](X)*t
            LIST[i].append(X[i])
    return LIST      
    
```

```python
def f1(X):
    m = 1.0
    return X[1]/m
def f2(X):
    k = 1.0
    return -k*X[0]
```

```python
X = [1.0,0.1]
F = [f1,f2]
t = 0.05
n = 150
```

```python
list_pts = Euler_gen(X,F,t,n)
```

```python
time = np.arange(1.0,1.0+n*t,t)
```

```python
plt.plot(time,list_pts[0])
plt.plot(time,list_pts[1])
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**3 - 0.165*x**2 + 0.0003993
def f_der(x):
    return 3*x**2 - 0.33*x
def nr(f,x,e):
    i = 0
    while(f(x)>e):
        x = x - f(x)/f_der(x)
        i = i+1
    return x,i
def nr_no_error(f,a,x):
    for j in range(50):
        x = x - g(x,a)/g_der(x)
    return x
def bisection(a,b,f,epsilon):
    error = epsilon+1
    iterations = 0
    while(abs(error)>epsilon):
        iterations = iterations + 1
        if f(a)*f(b)<0:
            #bisection method
            c = (a+b)/2
            if f(a)*f(c)<0:
                b=c
            else:
                a=c
            error = f(c)
           
        if f(a)*f(b)>0:
            #no root in this interval
            return None,None
    return c,iterations
```

```python
x = np.arange(-0.2,0.2,0.01)
y_a = f(x)
y_b = f_der(x)
plt.plot(x,y_a)
plt.plot(x,y_b)
plt.plot(x,0*x)
plt.plot(0.11, f_der(0.11), marker="o")
plt.plot(0.0, f_der(0), marker="o")
plt.grid()
```

```python
print(nr(f,0.05,0.0001))
print(nr(f,0.11,0.0001))  #0.11 is a point of minima, therefore f_der(x) is 0 and root cannot be calculated
# using 0 will also give no root as that is point of maxima
```

```python
e_nr = np.logspace(-1,-20,100)
y = []
for error in e_nr:
    y.append(nr(f,10,error)[1])
plt.plot(e_nr,y)
plt.grid()
```

```python
print(bisection(0.02,0.1,f,0.0001))
print(bisection(0.11,0.2,f,0.0001))
print(bisection(1.8,2.5,f,0.0001))
```

```python
e_bis = np.logspace(-1,-20,100)
y_2 = []
for error in e_bis:
    y_2.append(bisection(0,0.1,f,error)[1])
plt.plot(e_bis,y_2)
plt.grid()
plt.xscale('log')
plt.yscale('log')
```

```python
plt.plot(e_bis,y)
plt.plot(e_nr,y_2)
plt.grid()
plt.xscale('log')
plt.yscale('log')
```

```python
def g(x,a):
    return (1/x) - a
def g_der(x):
    return -1/(x**2)
```

```python
print(nr_no_error(g,3.0, 0.6))
```

```python
def nr_no_err_plot(f,a,x):
    X = []
    Y = []
    rprcl = 1/a
    for j in range(50):
        x = x - g(x,a)/g_der(x)
        X.append(j)
        Y.append(abs(rprcl-x))
    plt.plot(X,Y)
    plt.grid()
```

```python
print(nr_no_err_plot(g,3.0, 0.6))
```

```python
def diagonal_dominance_check(A):
    counter = 0
    for i in range (A.shape[1]):
        sum_row = 0
        for j in range (A.shape[1]):
            if (i!=j):
                sum_row = sum_row + abs(A[i][j])
        if sum_row >= abs(A[i][i]):
            print("Not strictly diagonally dominant at row ", i)
            counter = counter+1
    if (counter==0):
        print("Matrix is diagonally dominant")
```

```python
A = np.array([[-6,2,1,2,1],[3,8,-4,1,0],[-1,1,4,10,1],[3,-4,1,9,2],[2,0,1,3,10]])
diagonal_dominance_check(A)
```

```python
B = np.array([[18,3,6,-3],[9,13,-5,2],[-3,-2,4,9],[6,0,11,3]])
diagonal_dominance_check(B)
```

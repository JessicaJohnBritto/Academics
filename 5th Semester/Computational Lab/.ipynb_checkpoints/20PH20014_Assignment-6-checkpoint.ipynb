```python
import numpy as np
import matplotlib.pyplot as plt
```

# ODE-1

```python
y=4
h=0.05
N=60

def f(x,y):
    return (((2*y)-(18*x))/(1+x))

def euler(x, y, h, N):
    e=np.linspace(0,3, num=N)
    X, Y = [], []
    for i in e:
        y_new = y + (h*(f(i,y)))
        y = y_new
        X.append(i)
        Y.append(y)
    return plt.plot(X,Y, color="red")

def analytic_function(x):
    return ((-5*(x**2))+(8*x)+4)

def analytical(N):
    e=np.linspace(0,3, num=N)
    X, Y = [], []
    for i in e:
        X.append(i)
        Y.append(analytic_function(i))
    return plt.plot(X,Y, color="blue")

# Rung-Kutta Method 2nd order
def rk2(x, y, h, N):
    e=np.linspace(0,3, num=N)
    X, Y = [], []
    for i in e:
        k1 = h*f(i,y)
        k2 = h*f(i+h/2, y+k1/2)
        y_new = y + k2
        y = y_new 
        X.append(i)
        Y.append(y)
    return plt.plot(X,Y, color="green")
```

```python
analytical(60)
euler(0, 4, 0.05, 60)
rk2(0,4,0.05,60)
```

```python
def error(N):
    e=np.linspace(0,3, num=N)
    X, Y =[],[]
    y=4
    for i in e:
        y_n = y + (h*(f(i,y)))
        y = y_n
        y_a = analytic_function(i)
        error = (abs(y_a-y_n))
        X.append(i)
        Y.append(error)
    return plt.plot(X,Y, color="purple")
error(60)
```

# Coupled ODE - 1

```python
def gg(x,p):
    m=1
    return (p/m)
def hh(x,p):
    k=1
    return (-k*x)
def C_ODE(X, t):
    p=X[1]
    x=X[0]
    time = np.arange(0,20*(np.pi), t)
    XX, Y, L = [], [], []
    for i in time:
        p_n = p + t*(hh(x,p))
        x_n = x + t*(gg(x,p))
        p = p_n
        x = x_n
        XX.append(x)
        Y.append(p)
        plt.plot(XX,Y)
    return XX,Y, time
X = [1.,0.1]
```

```python
XX1, Y1, t1 = C_ODE(X, 2*(np.pi)/10)
plt.figure()
plt.plot(t1, XX1, color = "orange")
plt.plot(t1, Y1, color = "yellow")
```

```python
XX2, Y2, t2 = C_ODE(X, 2*(np.pi)/100)
plt.figure()
plt.plot(t2, XX2, color = "orange")
plt.plot(t2, Y2, color = "yellow")
```

```python
XX3, Y3, t3 = C_ODE(X, 2*(np.pi)/1000)
plt.figure()
plt.plot(t3, XX3, color = "orange")
plt.plot(t3, Y3, color = "yellow")
```

```python
# Analytical value of Energy
E = 1/2
def EEn(XX, Y, t):
    E=[]
    for i in range(len(XX)):
        En = 0.5*((XX[i])**2 + (Y[i])**2)
        error = (En-E)/E
        E.append(error)
    return plt.plot(E, t, color = "purple")

EEn(XX1, Y1, t1)
plt.figure()
EEn(XX2, Y2, t2)
plt.figure()
EEn(XX3, Y3, t3)
```

# ODE-1

```python
y=4
h=0.05
N=60

def f(x,y):
    return (((2*y)-(18*x))/(1+x))

def euler(x, y, h, N):
    e=np.linspace(0,3, num=N)
    X, Y = [], []
    for i in e:
        y_new = y + (h*(f(i,y)))
        y = y_new
        X.append(i)
        Y.append(y)
    return plt.plot(X,Y, color="red")

def analytic_function(x):
    return ((-5*(x**2))+(8*x)+4)

def analytical(x,h,N):
    e=np.linspace(x, x+h*N, N+1)
    X, Y = [], []
    for i in e:
        X.append(i)
        Y.append(analytic_function(i))
    plt.plot(X,Y, color="blue")
    return(Y)

# Rung-Kutta Method 2nd order
def rk2(x, y, h, N):
    e=np.linspace(0,3, num=N)
    X = []
    Y = []
    for i in e:
        Y.append(y)
        k1 = h*f(i,y)
        k2 = h*f(i+ h*0.5, y+ k1*0.5)
        y_new = y + k2
        y = y_new 
        X.append(i)
    plt.plot(X,Y, color="green")
    return (Y)
```

```python
analytical(0, 0.05, 60)
euler(0, 4, 0.05, 60)
Y=rk2(0,4,0.05,60)
```

```python
# Euler
def error(N):
    h=0.05
    e=np.linspace(0,3, num=N)
    X, Y =[],[]
    y=4
    for i in e:
        y_n = y + (h*(f(i,y)))
        y = y_n
        y_a = analytic_function(i)
        error = (abs(y_a-y_n))
        X.append(i)
        Y.append(error)
    print(Y)
    return plt.plot(X,Y, color="purple")
error(60)
```

```python
# Runge-Kutta

def f(x,y):
    return (((2*y)-(18*x))/(1+x))

def analytic_function(x):
    return ((-5*(x**2))+(8*x)+4)

def error1(N):
    h = 0.05
    e=np.linspace(0,3, num=N)
    X, Y =[],[]
    y=4
    for i in e:
        k1 = h*f(i,y)
        k2 = h*f(i+ h*0.5, y+ k1*0.5)
        y_new = y + k2
        y_a = analytic_function(i)
        error = (abs(y_a-y_new))
        Y.append(error)
        y = y_new
    print(Y)
    return plt.plot(e,Y, color="green") 
error1(60)
```

# Coupled ODE - 1

```python
# Part - 1

def gg(x,p):
    m=1
    return (p/m)
def hh(x,p):
    k=1
    return (-k*x)
def C_ODE(X, t, hh, gg):
    p=X[1]
    x=X[0]
    time = np.arange(0,20*(np.pi), t)
    XX, Y, L = [], [], []
    for i in time:
        p_n = p + t*(hh(x,p))
        x_n = x + t*(gg(x,p))
        p = p_n
        x = x_n
        XX.append(x)
        Y.append(p)
        plt.plot(XX,Y)
    return XX,Y, time
X = [1.,0.1]
```

```python
XX1, Y1, t1 = C_ODE(X, 2*(np.pi)/10, hh, gg)
plt.figure()
plt.plot(t1, XX1, color = "orange")
plt.plot(t1, Y1, color = "yellow")
```

```python
XX2, Y2, t2 = C_ODE(X, 2*(np.pi)/100, hh, gg)
plt.figure()
plt.plot(t2, XX2, color = "orange")
plt.plot(t2, Y2, color = "yellow")
```

```python
# Analytical Values of x and p from equation - x^2 + p^2 = c
# Parameterising circle

def analytical_xp(x,p, t):
    c = (p**2) + (x**2)
    return (c*np.cos(t), c*np.sin(t))

def C_ODE2(X, t, analytical_xp):
    p=X[1]
    x=X[0]
    time = np.arange(0,20*(np.pi), t)
    XX, Y, L = [], [], []
    for i in time:
        l = analytical_xp(x, p, i)
        XX.append(l[0])
        Y.append(l[1])
        plt.plot(XX,Y)
    return XX,Y, time

X = [1.,0.1]
XX, Y, t = C_ODE2(X, 2*(np.pi)/10, analytical_xp)
plt.plot(XX, Y)
plt.figure()
plt.plot(t, XX, color = "orange")
plt.plot(t, Y, color = "yellow")
```

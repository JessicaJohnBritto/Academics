# Problem-1
For double precision format

```python
s = 1.0
```

```python
x, y=[], []
for k in range(1, 100, 1):
    s = 0.5*s
    t = s+1.0
    x.append(k)
    y.append(t)
    if(t <= 1.0):
        s = 2.0*s
        print(k-1)
        print("\n")
        print(s)
        print("\n")
        break
plt.plot(x,y)
```

# Problem-1
For single precision format

```python
s = np.float32(1.0)
x, y=[], []
for k in range(1, 100, 1):
    s = np.float32(0.5)*s
    t = s+np.float32(1.0)
    x.append(k)
    y.append(t)
    if(t <= 1.0):
        s = np.float32(2.0)*s
        print(k-1)
        print("\n")
        print(s)
        print("\n")
        break
plt.plot(x,y)
```

## Problem 2

```python
######### y=(x^2 + 1.0)^(0.5)-1.0
# 1 -> y <-  (x^2 + 1.0)^(0.5)-1.0    A
# 2 ->  y <-    x^2/[(x^2+1.0)^(0.5)+1.0]      B
import math
k, m, A, B = [], [], [], []
for i in range(1,16,1):
    x = 10**(-i)
    a = (math.sqrt((x**2)+1.0)) - 1.0
    A.append(a)
    b = (x**2)/((math.sqrt((x**2)+1.0)) + 1.0)
    B.append(b)
    dev = abs((a-b)/b)
    k.append(x)
    m.append(dev)
plt.plot(k, m)
plt.yscale('log')
plt.xscale('log')
```

```python
print(A)
print('\n')
print(B)
```
